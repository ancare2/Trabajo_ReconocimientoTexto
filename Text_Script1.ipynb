```python
import numpy as np
import matplotlib.pyplot as plt

import pywt
import pywt.data
from PIL import Image
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
imagen_matrix = digits.images[0]
repres = Image.fromarray(np.uint8(imagen_matrix))
repres.show()
```

```python

# Load image
original = Image.open("data\images\Hola_Mundo.jpeg")
```

```python
grayscale_original = original.convert("L")
imagen_matrix = np.array(grayscale_original)
len(imagen_matrix)
```

```python

# Wavelet transform of image, and plot approximation and details
titles = ['Approximation', ' Horizontal detail',
          'Vertical detail', 'Diagonal detail']
coeffs2 = pywt.dwt2(imagen_matrix, "haar")
LL, (LH, HL, HH) = coeffs2
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()
plt.show()
```
